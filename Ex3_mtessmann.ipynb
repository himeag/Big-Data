## TASK 13

```python
# obtain the img in red, green and blue scale and print them
img_png_red = img_png.copy()
img_png_green = img_png.copy()
img_png_blue = img_png.copy()

img_png_red[:,:,[1,2]] = 0
img_png_green[:,:,[0,2]] = 0
img_png_blue[:,:,[0,1]] = 0

stacked = np.hstack((img_png_red, img_png_green, img_png_blue))

plt.imshow(stacked)
```

## Task 14

```python
# perform gamma correction on the image and display it
gamma = .2

# img_gamma_pt2 = img_png.copy()
img_gamma_pt2 = img_png[:,:,:3] ** gamma
img_gamma_5 = img_png[:,:,:3] ** 5
img_gamma_1 = img_png[:,:,:3] ** 1

stacked = np.hstack((img_gamma_1, img_gamma_5, img_gamma_pt2))
plt.imshow(stacked)
```

```python
# OPTIONAL - print 3 gamma corrected plots using subplots 
fig, ax = plt.subplots(1,3)
fig.suptitle('Horizontally stacked subplots')

ax[0].set_title('gamma = .2')
ax[1].set_title('gamma = .1')
ax[2].set_title('gamma = .5')

ax[0].set_yticks([])
ax[1].set_yticks([])
ax[2].set_yticks([])

ax[0].imshow(img_gamma_pt2)
ax[1].imshow(img_gamma_1)
ax[2].imshow(img_gamma_5)
```
