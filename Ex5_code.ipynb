# Task 0.

```python
import numpy as np
import pandas as pd
```

# Task 2. Read in data

```python
pop = pd.read_csv('popDF.csv', index_col=[0])
pop.head(5)
```

# Task 3. Fill missing values


## 3a. Use DF.fillna()


```python
pop['Regionfilled'] = pop['Region'].fillna(-1)
pop['Regionfilled'].unique()
```

## 3b. Replace (missing) values with specified values using  a mapping

```python
pop['Regionfilled'].replace({-1:"NoReg"}, inplace=True)
```

## 3c. Replace values by setting them equal to new value

```python
pop['RegionAll'] = pop['Region']
pop.loc[pop.index.isin(['Puerto Rico']), 'RegionAll'] = "PR"
pop.loc[pop.index.isin(['District of Columbia']), 'RegionAll'] = "DC"
pop['RegionAll']
```

## 3d. Display the values

```python
pop.loc[pop.index.isin(['District of Columbia', 'Puerto Rico']), ['Region', 'Regionfilled', 'RegionAll']]
```

# Task 4. Drop a column and view 5 randomly sampled rows

## 4a. drop column

```python
pop.drop('Regionfilled', axis=1, inplace=True, errors='ignore')
pop.head()
```

## 4b. Sample rows

```python
pop.sample(5)
```

# Task 5. Binning - Cuts and Quantiles


## 5a. Distribution of 'y2019'

```python
pop['y2019'].describe()
```

## 5b. Produce equal size bins

```python
pd.cut(pop.y2019, 5).head(10)
```

### 5b.i. How many states/territories lie in each bin?

```python
pd.cut(pop.y2019, 5).value_counts()
```

## 5c. Produce quantiles - with values equally distributed across bins

```python
pd.qcut(pop.y2019, 5).tail(10)
```

### 5c.i. How many states/territories lie in each bin?

```python
pd.qcut(pop.y2019, 5).value_counts()
```

# Task 6. Create dummy variables (one hot encoding)

## 6b. Join two DataFrames with common keys using DF. join()

# Task 7. Create a new DF with subset of the columns and rename its index 

## 7a. Select a subset of columns

```python
popDR_3y = pop.loc[:,['Region', 'RegionAll','y2017', 'y2018', 'y2019']].copy()
popDR_3y.head()
```

## 7b. Give the index a name

```python
popDR_3y.index.rename('State', inplace=True)
popDR_3y.head(3)
```

# Task 8. Applying functions to DataFrame columns

## 8a. Using Series unary and binary ufuncs - elementwise calculations

### Which state has highest/lowest popchange % from 2018 to 2019?

```python
popDR_3y['popchangepct']=(popDR_3y.y2019-popDR_3y.y2018)/popDR_3y.y2018
popDR_3y.head()
```

```python
popDR_3y[['y2018', 'y2019', 'popchangepct']].sort_values(ascending=False, by='popchangepct').head(5)
```

## 8b. Using DF.applymap()


### Express pop in millions, rounded to 2 digits

```python
popDR_3y.loc[:, ['y2017', 'y2018', 'y2019']].applymap(lambda x: np.round(int(x)/1000000,2)).sort_index().head()
                                                      
                                                      
```

## 8c. Using DF.apply()

### 8c.i. Using apply() to accces column values -  Standardization of a variable 

```python
popDR_3y[['y2017','y2018','y2019']].apply(lambda x: ( x - x.mean() ) / x.std() ).describe()

```

### 8c.ii. Using apply() to accces column values -  Normalization of a variable 

```python
popDR_3y[['y2017','y2018','y2019']].apply(lambda x: ( x - x.min()) /(x.max()-x.min())).describe()
```

### 8c.iii. Using apply() to access row values 

```python
# popDR_3y.loc[:,['y2018', 'y2019']].apply(lambda row: row['y2018'] > row['y2019'] , axis=1).value_counts()
popDR_3y.apply(lambda row: row['y2018'] > row['y2019'] , axis=1).value_counts()

```

# Task 9. Perform calculations over groups using DF.groupby()

## 9a. Computing within-groups stats such as sum

```python
popDR_3y[['y2017', 'y2018', 'y2019']].groupby(popDR_3y['RegionAll']).sum().sort_index()

```

## 9b. computing standardized variables within each group rather than for the whole data

```python
popDR_3y[['y2017', 'y2018', 'y2019']]\
    .groupby(popDR_3y['RegionAll'])\
    .apply(lambda x: ( x - x.mean() ) / x.std() )\
    .sort_index().head()
```

```python
# see the difference with standardization without grouping

popDR_3y[['y2017', 'y2018', 'y2019']]\
    .apply(lambda x: ( x - x.mean() ) / x.std() )\
    .sort_index().head()
```

## 9c. Filling rows with transform()

```python
popDR_3y['y2019'].groupby(popDR_3y['RegionAll']).transform('count')
popDR_3y['y2019'].groupby(popDR_3y['RegionAll']).transform(lambda x: x.count())

```

## 9d. Grouping on the index rather than a regular column(s)

```python
popDR_3y.reset_index().groupby('State').mean()
```

# Task 10. Re-shaping DataFrames

## 10a. Stacking


```python
popDR_3y['RegionMean']= popDR_3y['y2019'].groupby(popDR_3y['RegionAll']).transform('mean')

popDR_3y.head()

popDR_3y_stack = popDR_3y[['y2018', 'y2019']].stack()
popDR_3y_stack.index
```

## 10b. Unstacking

```python
popDR_3y_stack.unstack()
```
