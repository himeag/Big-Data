# Task 11

```
# #11a - groupBy()

GtDeviceGroups = streamingDF.groupBy("Device","gt").count()

GtDeviceGroupsQuery = GtDeviceGroups.writeStream\
  .outputMode('complete')\
  .queryName('gt_device_groups_results')\
  .format('memory')\
  .start()

for x in range(5):
  spark.sql("SELECT * FROM gt_device_groups_results").show()
  sleep(5)

GtDeviceGroupsQuery.stop()
```

```
# #11a - rollup()

GtDeviceRollup = streamingDF.rollup("Device","gt").count()

GtDeviceRollupQuery = GtDeviceRollup.writeStream\
  .outputMode('complete')\
  .queryName('gt_device_rollup_results')\
  .format('memory')\
  .start()

for x in range(5):
  spark.sql("SELECT * FROM gt_device_rollup_results").show()
  sleep(5)

GtDeviceRollupQuery.stop()
```

```
# #11a - cube()

GtDeviceCube = streamingDF.cube("Device","gt").count()

GtDeviceCubeQuery = GtDeviceCube.writeStream\
  .outputMode('complete')\
  .queryName('gt_device_cube_results')\
  .format('memory')\
  .start()

for x in range(5):
  spark.sql("SELECT * FROM gt_device_cube_results").show()
  sleep(5)

GtDeviceCubeQuery.stop()
```

```
# #12a

stairsonlyDF = streamingDF.filter(F.col('gt').contains('stairs'))\
                      .filter(F.col('Device')=='nexus4_1')\
                      .filter((F.col('User')=='a') | (F.col('User')=='b'))\
                      .groupby('User')\
                      .count()

StairsOnlyQuery = stairsonlyDF.writeStream\
  .outputMode('complete')\
  .queryName('stairs_only_results')\
  .format('memory')\
  .start()
```

#Task 13

```
# #13a
gtmodelstatsDF = streamingDF.groupBy('gt', 'Model')\
        .agg(F.avg('x'), F.avg('y'), F.avg('z'), F.count('Device'))\
        .withColumnRenamed('avg(x)', 'x_avg')\
        .withColumnRenamed('avg(y)', 'y_avg')\
        .withColumnRenamed('avg(z)', 'z_avg')\
        .withColumnRenamed('count(Device)', 'devicecount')

```

```
# 13b
gtmodelstatsQuery = gtmodelstatsDF.writeStream\
  .outputMode('complete')\
  .queryName('gt_model_stats_results')\
  .format('memory')\
  .start()

for x in range(5):
  spark.sql("SELECT * FROM gt_model_stats_results WHERE x_avg>0").show()
  sleep(5)
  
```

```
# #13c
gtmodelstatsQuery.stop()
```

# Task 14

```
# 14a

historicalAvg = staticDF.groupBy('gt', 'Model')\
        .agg(F.avg('x'), F.avg('y'), F.avg('z'))\
        .withColumnRenamed('avg(x)', 'x_avg1')\
        .withColumnRenamed('avg(y)', 'y_avg1')\
        .withColumnRenamed('avg(z)', 'z_avg1')
```

```
historicalAvg.head(5)
```

```
# 14b

ComparingAvgDF = streamingDF.groupBy('gt', 'Model')\
        .agg(F.avg('x'), F.avg('y'), F.avg('z'))\
        .withColumnRenamed('avg(x)', 'x_avg2')\
        .withColumnRenamed('avg(y)', 'y_avg2')\
        .withColumnRenamed('avg(z)', 'z_avg2')\
        .join(historicalAvg, on=['gt', 'Model'])

ComparingAvgQuery = ComparingAvgDF.writeStream\
  .outputMode('complete')\
  .queryName('comparing_average_results')\
  .format('memory')\
  .start()

for x in range(5):
  sleep(30)
  spark.sql("SELECT * FROM comparing_average_results").show()
  

ComparingAvgQuery.stop()
                         
```

```
# 14c

#####
## The last three columns, appended with 1, are from the static historicalAvg dataframe. These averages are not being affected by incoming data and do not change - they're already calculated and static. In contrast, the averages being updated from the stream data, denoted by *_avg2, are being updated every loop and are therefore changing with every write out. 
```
