# 4. Data Cleaning

## 4a. test for nulls

```python
# any will return boolean value for each column
statesPop_df.isnull().any(axis=1)
```

```python
statesPop_df.isnull().all(axis=1).value_counts()
```

## 4b. count null and non-null vals

```python
statesPop_df.count()
```

## 4c. drop rows/cols that are completly null

```python
print(f"\n\nROW NAs\n{statesPop_df.isnull().any(axis=1)}")
```

```python
statesPop_df_dropNA = statesPop_df.dropna(axis=0, how='all')
statesPop_df_dropNA.tail(5)
statesPop_df_dropNA.shape
```

```python
statesPop_df_dropNA.isnull().any()
```

## 4d. check for duplicate and uniques values

```python
print(f"UNIQUE:\n\n{statesPop_df_dropNA.nunique()}")
print(f"\n\nDUPLUCATES:\n\n{statesPop_df_dropNA.duplicated()}")
print(f"\n\nCOUNT:\n{statesPop_df_dropNA.duplicated().value_counts()}")
```

## 4e. drop duplicate rows

```python
statesPop_df_na_dup = statesPop_df_dropNA.drop_duplicates(keep='first',
                                                         inplace=False)
print(f"\n\nCOUNT:\n{statesPop_df_na_dup.duplicated().value_counts()}")
```

```python
statesPop_df_na_dup.shape
```
